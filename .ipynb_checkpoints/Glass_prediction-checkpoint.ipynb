## Feature Transformation/Engineering

```python
train_data['log_max_luminosity'] = np.log(train_data['max_luminosity'])
train_data['log_pixel_area'] = np.log(train_data['pixel_area'])
```

```python
train_data['log_xrange'] = np.log1p(abs(train_data['xmax'] - train_data['xmin']))
train_data['sqrt_yrange'] = np.sqrt(abs(train_data['ymax'] - train_data['ymin']))
```

```python
train_data['max_luminosity/thickness'] = train_data['max_luminosity']/train_data['thickness']
```

```python

train_data['avg_x_component'] = (train_data['x_component_1'] + train_data['x_component_2'] + train_data['x_component_3'] + train_data['x_component_4'] + train_data['x_component_5'])/5
```

```python

train_data = train_data.drop(['grade_A_Component_2','log_area'], axis = 1)
```

```python
train_data = train_data.drop(['total_x'], axis = 1)
```

## Dependent and Independent variable seperation

```python
x = train_data.drop('class', axis=1)
y = train_data['class']
x.shape, y.shape
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y, test_size=0.3,random_state=101, stratify=y)
```

### Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_x = sc.fit_transform(train_x)
test_x = sc.transform(test_x)
```

```python
#importing random forest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, auc
from sklearn.metrics import roc_auc_score
```

```python
clf = RandomForestClassifier(n_estimators=15, random_state=100,criterion='entropy')
```

```python
clf.fit(train_x,train_y)
```
